```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/hanwoolJeong/lectureUniv/main/testData_LogisticRegression.txt", sep="\s+")
x = np.array(df.values[:,0])
y_raw = np.array(df.values[:,1])
plt.plot(x, y_raw, "k.")
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x_test = np.linspace(-10, 10, num=101)
plt.plot(x_test, sigmoid(x_test), "k-")
```

```python
N = len(x)
x_bias = np.c_[np.ones([N, 1]), x].T
y = y_raw.reshape(N, 1)
X = x_bias.T
```

**NLL의 Gradient**
$$
X^T (\mu - y)
$$


```python
eta = 0.1 # learning rate
n_iterations = 1000
wGD = np.zeros([2, 1]) # initialize w to 0
wGDbuffer = np.zeros([2, n_iterations+1])

for iteration in range(n_iterations):
    mu = sigmoid(wGD.T.dot(x_bias)).T
    gradients = X.T.dot(mu-y)
    wGD -= eta*gradients
```

```python
X_test = np.linspace(0, 10, num=N).reshape(N, 1)
X_test_bias = np.c_[np.ones([N, 1]), X_test]
a = sigmoid(wGD.T.dot(X_test_bias.T))
plt.plot(X_test, a.T, "r-")
plt.plot(x, y_raw, "k.")
```
